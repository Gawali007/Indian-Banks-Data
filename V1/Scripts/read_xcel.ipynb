```python
import os
import json
from collections import OrderedDict
from itertools import islice
from openpyxl import load_workbook
blank_space=''
def checkTuple(row,index):
    if index < len(row):
        if row[index]!=None:
            return row[index]
        else:
            return blank_space
    else:
        return blank_space
    
def checkFolderIfExists(folder_name):
    if not os.path.exists(folder_name):
        os.makedirs(folder_name)
        print("folder created :" +folder_name)
        
        

def ExcelToJson(file_name):        
    wb = load_workbook(file_name)
    sheet = wb[wb.sheetnames[0]]
    json_file_name=""
    banks_list = []
    for row in islice(sheet.values, 1, sheet.max_row):
        banks = OrderedDict()
        #checkFolderIfExists(str(checkTuple(row,0)))
        banks['bank'] = checkTuple(row,0)
        banks['ifsc'] = checkTuple(row,1)
        banks['micr'] = checkTuple(row,2)
        banks['branch'] = checkTuple(row,3)
        banks['address'] = checkTuple(row,4)
        banks['city1'] = checkTuple(row,5)
        banks['city2'] = checkTuple(row,6)
        banks['state'] = checkTuple(row,7)
        banks['stdcode'] = checkTuple(row,8)
        banks['phone'] = checkTuple(row,9)
        banks_list.append(banks)
    json_file_name="test/"+banks['bank'].replace(' ','_')+".json"
    print(json_file_name)
    j = json.dumps(banks_list, indent = 1)
    os.rename(file_name,banks['bank'].replace(' ','_')+".xlsx")
    with open(json_file_name, 'w') as f:
        f.write(j)
        
        


for root, dirs, files in os.walk("C:\\Users\\Kaustubh\\Downloads\\"):
    for file in files:
        if file.endswith('.xlsx'):
            print(file)
            ExcelToJson(file)
            

```
